```python
# Read the image below, apply the masks to detect horizontal, vertical and diagonal lines.

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('img.png')

horizontal_mask = np.array([[-1,-1,-1], [2,2,2], [-1,-1,-1]])
vertical_mask = np.array([[-1,2,-1], [-1,2,-1], [-1,2,-1]])
neg45_mask = np.array([[2,-1,-1], [-1,2,-1], [-1,-1,2]])
pos45_mask = np.array([[-1,-1,2], [-1,2,-1], [2,-1,-1]])

horizontal = cv2.filter2D(img, -1, horizontal_mask)
vertical = cv2.filter2D(img, -1, vertical_mask)
diagonal1 = cv2.filter2D(img, -1, neg45_mask)
diagonal2 = cv2.filter2D(img, -1, pos45_mask)

plt.subplot(2, 2, 1)
plt.imshow(horizontal)
plt.title('Horizontal Lines')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(vertical)
plt.title('Vertical Lines')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(diagonal1)
plt.title('-45 Lines')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(diagonal2)
plt.title('+45 Lines')
plt.axis('off')
```

```python
# Use a Sobel Filter to display horizontal, vertical and combined edges.

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('img1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
blur_img = cv2.GaussianBlur(gray_img, (3,3), 0)

sobel_x = cv2.Sobel(blur_img, cv2.CV_64F, dx=1, dy=0, ksize=5)
sobel_y = cv2.Sobel(blur_img, cv2.CV_64F, dx=0, dy=1, ksize=5)
sobel_xy = cv2.Sobel(blur_img, cv2.CV_64F, dx=1, dy=1, ksize=5) 

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(sobel_x, cmap = "gray")
plt.title('Horizontal Edges')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(sobel_y, cmap = "gray")
plt.title('Vertical Edges')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(sobel_xy, cmap = "gray")
plt.title('Combined Edges')
plt.axis('off')
```
