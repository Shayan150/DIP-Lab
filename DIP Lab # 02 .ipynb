```python
# TASK no. 06.

import numpy as np
img = cv2.imread('image 2.jpg')  
blue, green, red = cv2.split(img)
gray_img = 0.299 * red + 0.587 * green + 0.114 * blue
gray_img = gray_img.astype(np.uint8)
cv2.imshow('Grayscale Image', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
